## 데이터 전처리

### 인코딩

````python
# 프라임 회원 유무 인코딩 (1 = 프라임 회원, 0 = 일반 회원)
map_prime = {'N':0, 'Y':1}
df['prime_yn'] = df['prime_yn'].map(map_prime)
df
````

### 이벤트 상품
product_name 앞에 [ ] 들어가면 1인 컬럼 만들기
--> 아마 나중에 따로 골라내야 할수도

````python
# 이벤트 할인 제품 총 구매 횟수 (event_product)
df['event_product'] = np.where(df['product_name'].str.contains(r'\['), 1, 0)
df
````

````python
# 이벤트 [] 안 내용 및 개수 확인
import pandas as pd
import re

def extract_words_within_brackets(text):
    pattern = r'\[(.*?)\]'  # "[]" 사이의 내용을 추출하기 위해 정규 표현식 패턴 지정
    matches = re.findall(pattern, text)
    if matches:
        words = ' '.join(re.findall(r'\b\w+\b', matches[0]))  # 내부 단어만 추출 후 공백으로 구분하여 문자열 생성
        return words
    return ''

# 일반 회원에 [] 포함된 내부 단어 컬럼 추출
df_event = df['product_name'].apply(extract_words_within_brackets)
df_event = df_event.dropna()
#pd.set_option('display.max_rows', None)
print(df_event.value_counts())
````

### 제품별 카테고리 - 크롤링 데이터 활용

````python
category = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BDA 5기/2023 BDA 공모전/crawling_cj.csv', encoding = 'cp949')
category
````

````python
category.drop('Unnamed: 0', axis=1, inplace=True)
category
````

````python
category['category'].value_counts()
````

````python
def assign_category(df, category):
    df = df.copy()
    merged_df = pd.merge(df, category, how='left', left_on='product_name', right_on='item')
    df['category'] = merged_df['category']
    return df

# df에 category 넣기
df = assign_category(df, category)
df
````

````python
# NaN 개수 확인 --> 꽤 많음.. 다 삭제?
nan_count = df['category'].isna().sum()
print(nan_count)
````

### 주문번호별로 합쳐서 새로운 df 만들기
- 주문번호(scd)별로 합치면 제품명 컬럼은 삭제됨
- 위에서 추가했던 category 변수는 고유값으로 부여함

````python
# df_2의 product_name은 구매한 제품 고유 개수임!
df_2 = df.groupby('scd').agg({'product_name':'count','category':'count'}).reset_index()
df_2
````

````python
df_2= df.groupby(['scd']).agg({'gender': 'first', 'age_grp':'first', 'employee_yn':'first',
                               'prime_yn':'first', 'weekday':'first', 'weekday2':'first',
                               'net_order_qty': 'sum', 'net_order_amt': 'sum', 'event_product': 'sum',
                               'product_name':'count', 'category':'count'}).reset_index()
df_2
````
